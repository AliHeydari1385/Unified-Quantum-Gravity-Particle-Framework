```python
# Step 2: Dynamic file finding
def find_file(pattern):
    for root, dirs, files in os.walk('/content'):
        for file in files:
            if pattern in file:
                return os.path.join(root, file)
    raise FileNotFoundError(f"File containing '{pattern}' not found. Check with !ls -R /content")

# Load CMB data
cl_file = find_file('ACT+SPT_cl.dat')
cov_file = find_file('ACT+SPT_cov.dat')

# Custom loader for cov.dat
def load_cov_custom(filepath):
    data = []
    with open(filepath, 'r') as f:
        for line in f:
            line = line.strip().replace('|', '').replace('---', '').replace('#', '')
            if line:
                try:
                    row = [float(x) for x in line.split() if x.strip()]
                    if row:
                        data.append(row)
                except ValueError:
                    pass
    data_flat = np.concatenate(data)
    n = int(np.sqrt(len(data_flat)))
    if n * n != len(data_flat):
        raise ValueError(f"Cov matrix not square: {len(data_flat)} elements, sqrt={n}")
    cov = np.array(data_flat).reshape(n, n)
    return cov

cmb_cl = np.loadtxt(cl_file)  # columns: l, Cl, err (89 lines)
cmb_cov = load_cov_custom(cov_file)  # 89x89
cmb_l = cmb_cl[:, 0]
cmb_cl_obs = cmb_cl[:, 1]
cmb_cl_err = cmb_cl[:, 2]

# Load SNIa data (fixed: handle header, convert to float)
snia_file = find_file('ES_AND_COVARPantheon%2BSH0ES.dat.txt')
snia_data = pd.read_csv(snia_file, sep='\s+', on_bad_lines='skip', header=0)  # header=0 to skip first line as header

# Select columns and convert to float (handle any non-numeric)
z_snia = pd.to_numeric(snia_data['zHD'], errors='coerce').dropna().values
mu_obs = pd.to_numeric(snia_data['MU_SH0ES'], errors='coerce').dropna().values
mu_err = pd.to_numeric(snia_data['MU_SH0ES_ERR_DIAG'], errors='coerce').dropna().values

# Ensure same length (drop any NaN rows)
min_len = min(len(z_snia), len(mu_obs), len(mu_err))
z_snia, mu_obs, mu_err = z_snia[:min_len], mu_obs[:min_len], mu_err[:min_len]

# Subsample for speed (optional: comment out for full ~1701 points)
z_snia, mu_obs, mu_err = z_snia[:500], mu_obs[:500], mu_err[:500]

print(f"Loaded CMB: {len(cmb_l)} points")
print(f"Loaded SNIa: {len(z_snia)} points (all floats)")

# Step 3: Define UQGPF v2.0 model (unchanged)
c = 3e8  # m/s
H0_base = 70  # km/s/Mpc
l_p = 1.616e-35
theta_s = 0.01
ell_string = 10000

def uqgpf_cmb_cl(ell, params):
    Omega_m, h, gamma, f_a, m_a, lambda_s = params
    alpha = 1.5
    beta = 0.3
    term1 = alpha * (ell * (ell + 1)) / (2 * np.pi) * Omega_m**beta * np.exp(-gamma * ell / 2500)
    term2 = (m_a / f_a) * np.sin(ell * theta_s)**2
    term3 = lambda_s * (l_p**2 / ell**2) * np.cos(ell / ell_string)
    return term1 + term2 + term3

def hubble_z(z, params):
    Omega_m, h, gamma, *_ = params
    H0 = h * H0_base
    Omega_L = 1 - Omega_m
    def integrand(zp):
        return c / (H0 * np.sqrt(Omega_m * (1 + zp)**3 + Omega_L + gamma * l_p**2 * (1 + zp)**(-2)))
    dl, _ = quad(integrand, 0, z)
    return (1 + z) * dl

def uqgpf_mu(z, params):
    dl = hubble_z(z, params)
    return 5 * np.log10(dl) + 25

# Step 4: Priors and Likelihood (unchanged)
def log_prior(theta):
    Omega_m, h, gamma, f_a, m_a, lambda_s = theta
    if (0.2 < Omega_m < 0.3 and 0.67 < h < 0.73 and 0.23 < gamma < 0.25 and
        9e11 < f_a < 1.1e12 and 9e-5 < m_a < 1.1e-4 and 0.3 < lambda_s < 0.7):
        return 0.0
    return -np.inf

def log_likelihood(theta, l_cmb, cl_obs, cov_cmb, z_snia, mu_obs, mu_err):
    cl_model = uqgpf_cmb_cl(l_cmb, theta)
    diff_cmb = cl_obs - cl_model
    inv_cov = np.linalg.inv(cov_cmb)
    chi2_cmb = diff_cmb.T @ inv_cov @ diff_cmb

    mu_model = np.array([uqgpf_mu(z, theta) for z in z_snia])
    chi2_snia = np.sum(((mu_obs - mu_model) / mu_err)**2)

    return -0.5 * (chi2_cmb + chi2_snia)

def log_probability(theta, *args):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, *args)

# Step 5: Run MCMC
ndim = 6
nwalkers = 32
nsteps = 50000

p0 = np.array([0.25, 0.70, 0.24, 1e12, 1e-4, 0.5]) + 1e-4 * np.random.randn(nwalkers, ndim)

args = (cmb_l, cmb_cl_obs, cmb_cov, z_snia, mu_obs, mu_err)
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=args)

try:
    sampler.run_mcmc(p0, nsteps, progress=True)
except KeyboardInterrupt:
    print("MCMC interrupted! Saving current state...")

samples = sampler.get_chain(discard=1000, thin=15, flat=True)
print(f"Acceptance fraction: {np.mean(sampler.acceptance_fraction)}")

best_params = np.median(samples, axis=0)
final_chi2 = -2 * log_likelihood(best_params, *args)
print(f"Final chi2: {final_chi2} (target ~8.5)")

# Step 6: Generate outputs (unchanged from previous)
# ... (copy the rest: corner plot, residuals CSV, LaTeX table, fit plot, ZIP)
labels = [r'\Omega_m', 'h', r'\gamma', r'f_a', r'm_a', r'\lambda_s']
fig = corner.corner(samples, labels=labels, truths=best_params)
fig.savefig('uqgpf_posteriors_v11.pdf')
display(FileLink('uqgpf_posteriors_v11.pdf'))

cl_model_best = uqgpf_cmb_cl(cmb_l, best_params)
res_cmb = cmb_cl_obs - cl_model_best
mu_model_best = np.array([uqgpf_mu(z, best_params) for z in z_snia])
res_snia = mu_obs - mu_model_best

residuals_df = pd.DataFrame({
    'cmb_l': cmb_l, 'cmb_res': res_cmb, 'cmb_err': cmb_cl_err,
    'snia_z': z_snia, 'snia_res': res_snia, 'snia_err': mu_err
})
residuals_df.to_csv('residuals_v11.csv', index=False)
display(FileLink('residuals_v11.csv'))

latex_table = r"""
\begin{table}[h]
\centering
\begin{tabular}{cc}
Parameter & Value \\
\hline
\Omega_m & %.3f \pm %.3f \\
h & %.3f \pm %.3f \\
\gamma & %.3f \pm %.3f \\
f_a & %.2e \pm %.2e \\
m_a & %.2e \pm %.2e \\
\lambda_s & %.3f \pm %.3f \\
\hline
\end{tabular}
\caption{Best-fit parameters for UQGPF v2.0}
\end{table}
""" % tuple(np.concatenate((best_params, np.std(samples, axis=0))))
with open('latex_table_v11.tex', 'w') as f:
    f.write(latex_table)
display(FileLink('latex_table_v11.tex'))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
ax1.errorbar(cmb_l, cmb_cl_obs, yerr=cmb_cl_err, fmt='o', label='Data')
ax1.plot(cmb_l, cl_model_best, label='UQGPF v2.0 Fit')
ax1.set_xlabel('$\ell$')
ax1.set_ylabel('$C_\ell$')
ax1.legend()
ax2.errorbar(z_snia, mu_obs, yerr=mu_err, fmt='o', label='Data')
ax2.plot(z_snia, mu_model_best, label='UQGPF v2.0 Fit')
ax2.set_xlabel('$z$')
ax2.set_ylabel('$\mu$')
ax2.legend()
fig.tight_layout()
fig.savefig('uqgpf_fit_v11.pdf')
display(FileLink('uqgpf_fit_v11.pdf'))

!zip -r Test_V2.2.zip uqgpf_posteriors_v11.pdf residuals_v11.csv latex_table_v11.tex uqgpf_fit_v11.pdf
display(FileLink('Test_V2.2.zip'))
```
