```python
# =========================
# Google Colab: UQCMF + Pantheon+ Subsample Fit
# =========================

# --- نصب وابستگی‌ها ---
!pip install emcee corner matplotlib numpy scipy

import numpy as np
import matplotlib.pyplot as plt
import emcee
import corner

# --- داده نمونه (۱۰ SN از Pantheon+؛ z, m_b_corr, err) ---
# این را می‌توانید با کل داده جایگزین کنید
data = np.array([
    [0.00122, 9.74571, 0.0991],
    [0.00122, 9.80286, 0.0971],
    [0.00256, 11.4703, 0.0399],
    [0.00256, 11.4919, 0.0931],
    [0.00299, 11.5227, 0.0567],
    [0.00317, 11.5416, 0.0732],
    [0.00345, 12.0048, 0.0864],
    [0.00360, 12.1025, 0.0896],
    [0.00370, 12.1441, 0.0944],
    [0.00382, 12.2234, 0.0988]
])
z, m_b, err = data[:,0], data[:,1], data[:,2]

# --- مدل UQCMF ---
def mu_UQCMF(theta, z):
    Om, h, beta = theta
    # فاصله لومینوسیتی با تقریب ساده FlatLCDM
    c = 299792.458 # km/s
    integral = z / np.sqrt(Om*(1+z)**3 + (1-Om))  # تقریبی، نه انتگرال دقیق
    D_L = (c / h) * (1+z) * integral
    mu_cosmo = 5 * np.log10(D_L) + 25
    Delta_mu_mind = beta * np.log(1 + z) / 1e-3
    return mu_cosmo + Delta_mu_mind

# --- توابع MCMC ---
def log_prior(theta):
    Om, h, beta = theta
    if 0 < Om < 1 and 60 < h < 80 and -0.01 < beta < 0.01:
        return 0.0
    return -np.inf

def log_likelihood(theta, z, m_b, err):
    model = mu_UQCMF(theta, z)
    return -0.5 * np.sum(((m_b - model) / err)**2)

def log_posterior(theta, z, m_b, err):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, z, m_b, err)

# --- پارامترهای اولیه ---
ndim, nwalkers = 3, 32
pos = [0.3, 70, 0.0] + 1e-4 * np.random.randn(nwalkers, ndim)

# --- اجرای MCMC ---
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_posterior, args=(z, m_b, err))
sampler.run_mcmc(pos, 2000, progress=True)

# --- پردازش خروجی ---
samples = sampler.get_chain(discard=200, flat=True)
labels = [r"$\Omega_m$", r"$h$", r"$\beta_{\rm mind}$"]

# نمودار Posteriors
fig = corner.corner(samples, labels=labels, truths=[np.median(samples[:,0]),
                                                    np.median(samples[:,1]),
                                                    np.median(samples[:,2])])
fig.savefig("posteriors_uqcmf.pdf")
fig.savefig("posteriors_uqcmf.png", dpi=300)
plt.show()

# چاپ نتایج میانه و عدم قطعیت
for i, lab in enumerate(labels):
    median = np.percentile(samples[:,i], 50)
    plus = np.percentile(samples[:,i], 84) - median
    minus = median - np.percentile(samples[:,i], 16)
    print(f"{lab} = {median:.4f} +{plus:.4f} -{minus:.4f}")

print("فایل‌های خروجی ذخیره شدند: posteriors_uqcmf.pdf و posteriors_uqcmf.png")
```
